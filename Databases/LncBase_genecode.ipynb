```python
from collections import OrderedDict

gencode = OrderedDict()

with open('gencode.v21.annotation.gtf') as f:
    for line in f:
        line = line.rstrip()
        if line.startswith('#'):
            continue
        lst = line.split('\t')
        if lst[2] == 'gene':
            geneInfo = lst[-1].split(';')
            geneId = geneInfo[0].split('"')[1].split('.')[0]
            geneName = geneInfo[4].split('"')[1]
            biotype = geneInfo[2].split('"')[1]
            #print (line)
            if geneId not in gencode.keys():
                gencode[geneId] = '\t'.join([geneName, biotype])
            
fh = open('gencode.v21.annotation.clean.txt', 'wt')

for ke, val in gencode.items():
    fh.write(ke+'\t'+val+'\n')
    
fh.close()
```
