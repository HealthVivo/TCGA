```python
########################################
######## GO

goMF = OrderedDict()
goBP = OrderedDict()
goCC = OrderedDict()


with open(databases+'ensembl_GO.txt') as f:
    for line in f:
        line = line.rstrip()
        lst = line.split('\t')
        
        if line.startswith('Gene'):
            continue
            
        elif len(lst) == 1:
            continue
            
        goterm = lst[2]+'\t'+lst[3]
            
        if lst[1] == 'molecular_function':
            if goterm not in goMF.keys():
                goMF[goterm] = []
            
            if lst[0] not in goMF[goterm]:
                goMF[goterm].append(lst[0])
                
        elif lst[1] == 'biological_process':
            if goterm not in goBP.keys():
                goBP[goterm] = []
            
            if lst[0] not in goBP[goterm]:
                goBP[goterm].append(lst[0])
                
        
        
        elif lst[1] == 'cellular_component':
            if goterm not in goCC.keys():
                goCC[goterm] = []
            
            if lst[0] not in goCC[goterm]:
                goCC[goterm].append(lst[0])
                
fh = open(databases+'ensembl_GO.cleaned.txt', 'wt')

for ke,val in goMF.items():
    fh.write('molecular_function'+'\t'+ke+'\t'+';'.join(val)+'\n')
    
for ke,val in goBP.items():
    fh.write('biological_process'+'\t'+ke+'\t'+';'.join(val)+'\n')
    
for ke,val in goCC.items():
    fh.write('cellular_component'+'\t'+ke+'\t'+';'.join(val)+'\n')
    
    
fh.close()
```

```python
########################################
######## GO


fh = open(databases+'GO/go-basic-02142017.obo.txt', 'wt')
fh.write('\t'.join(['Domain', 'goID', 'goName', 'Definition', 'is_a'])+'\n')


go = OrderedDict()
go['biological_process'] = []
go['molecular_function'] = []
go['cellular_component'] = []

switch = 0

with open(databases+'GO/go-basic.obo') as f:
    for line in f:
        line = line.rstrip()

        if line.startswith('[Term]'):
            if switch == 0:
                isA = []
                switch = 1
                continue
            else:
                go[goNamespace].append('\t'.join([goID, goName, goDef, ';'.join(isA)]))
                isA = []
            
        elif line.startswith('id:'):
            goID = re.sub('id:\s+', '', line)

        elif line.startswith('name:'):
            goName = re.sub('name:\s+', '', line)
            
        elif line.startswith('namespace:'):
            goNamespace = re.sub('namespace:\s+', '', line)
            
        elif line.startswith('def:'):
            goDef = re.sub('def:\s+', '', line)
            goDef = re.sub('"', '', goDef)
            
        elif line.startswith('is_a'):
            isA.append(re.search('GO:\d+',line).group(0))
            
        elif line.startswith('[Typedef]'):
            go[goNamespace].append('\t'.join([goID, goName, goDef, ';'.join(isA)]))
            break
            
            
            
for ke, val in go.items():
    for subv in val:
        fh.write(ke+'\t'+subv+'\n')
        

fh.close()
        
        
    
```
